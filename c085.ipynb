```python
def data(file,sheetname,outname,outsheetname):
    excel=xlrd.open_workbook(path+file,'rb')
    sheet=excel.sheet_by_name(sheetname)
    ID=[]
    FL=[]
    for r in range(sheet.nrows):#nrows有效行数
        ID.append(sheet.cell_value(r,0))#行，列
        FL.append(sheet.cell_value(r,1))
    AW=[]
    BW=[]
    CW=[]
    for c in range(2,sheet.ncols):#ncols:有效列数
            AN=0
            BN=0
            CN=0
            for r in range(1,sheet.nrows):
                if FL[r]=='A':
                    AN+=sheet.cell_value(r,c)
                if FL[r]=='B':
                    BN+sheet.cell_value(r,c)
                if FL[r]=='C':
                    CN+=sheet.cell_value(r,c)
            AW.append(AN)
            BW.append(BN)
            CW.append(CN)
    workbook=xlsxwriter.Workbook(path+outname)#建立文件
    worksheet=workbook.add_worksheet(outsheetname)
    worksheet.write(1,0,'A')
    worksheet.write(2,0,'B')
    worksheet.write(3,0,'C')
    for i in range(len(AW)):
        worksheet.write(0,i+1,i+1)
        worksheet.write(1,i+1,AW[i])
        worksheet.write(2,i+1,BW[i])
        worksheet.write(3,i+1,CW[i])
    workbook.close()
```

```python
def get_new_Z(name):
    cha=np.zeros(w)
    ObjL=np.zeros(w)
    newZ=np.zeros(w)
    for j in name:
        for z in range(w):
            ObjL[z]+=GHL[j][z]
    for z in range(w):
        cha[z]=C-ObjL[z]
    for z in range(w-1):
        if cha[z]<0:
            cha[z+1]+=cha[z]
            cha[z]=0
    if cha[w-1]<0:
        cha[w-1]=0
    s=np.sum(cha)
    for z in range(w):
        newZ[z]=cha[z]/s
    return newZ
```
